```python
first = True
errorCount = 0
for filename in os.listdir('/Users/jay_lohokare/Desktop/products'):
    try:
        fullName = '/Users/jay_lohokare/Desktop/products/' + filename
        with open(fullName) as f:
            d = json.load(f)
        try:
            if 'materials_products' in d['product']:
                del d['product']['materials_products']
            if 'compliance_sources' in d['product']:
                del d['product']['compliance_sources']
            if 'alternate_ids' in d['product']:
                del d['product']['alternate_ids']
        except:
            print "Some error deleting "
        df = json_normalize(d['product'])
        df['links_pricing_tiers_products'] = df['links.pricing_tiers_products']
        df['links_product_attachments'] = df['links.product_attachments']
        df['links_product_comments'] = df['links.product_comments']

        df.drop(columns='links.pricing_tiers_products', inplace=True)
        df.drop(columns='links.product_attachments', inplace=True)
        df.drop(columns='links.product_comments', inplace=True)
        
        with open('Desktop/products_all.csv', 'a') as csvFile:
            if first:
                first = False
                df.to_csv(csvFile, header=True)
            else:
                df.to_csv(csvFile, mode = 'a', header=False)
    except:
        errorCount += 1
        print str(errorCount) + " " + filename + " File is corrupt"
        eFile = open('Desktop/errorLog.txt','a') 
        eFile.write("Error found in " + filename + '\n') 
        eFile.close()
```
