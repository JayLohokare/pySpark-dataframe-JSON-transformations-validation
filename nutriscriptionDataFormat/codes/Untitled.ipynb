```python
import pandas as pd
import os
```

```python
df = pd.read_csv("all_fields_with_Region.csv", '|')
```

```python
for index, row in df.iterrows():
    try:
        branchName = str(row['BRANCH_NAME'])
        regionName = str(row['REGION_NAME'])
        divisionName = str(row['DIVISION_NAME'])
        growerId = str(row['ARS_GROWER_ID'])
        farmId = str(row['ARS_FARM_ID'])
        fieldId = str(row['ID'])

        fileName = regionName + '/' + divisionName + '/' + branchName + '/' + growerId + '/' + farmId + '/' + fieldId + '.json'
        jsonData =  row.to_json()

        try:
            os.makedirs(os.path.dirname(fileName))
        except:
            print "Ooops"
        obj = open(fileName, 'w')
        obj.write(jsonData)
        obj.close
    except:
        print "Oops agains"
        

    
```
