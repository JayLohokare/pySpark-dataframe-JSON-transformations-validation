```python
import pandas as pd
from collections import defaultdict
import json
import os
```

```python
df = pd.read_csv("all_fields_with_Region.csv", '|')
```

```python
df2 = df.filter(['FarmName', 'ARS_FARM_ID'], axis=1)
df2 = df2.rename(columns={'ARS_FARM_ID':'farm_id'})
```

```python
dfGrouped = df.filter(['ID', 'ARS_FIELD_NAME', 'ARS_FARM_ID' ]).groupby(['ARS_FARM_ID']).apply(lambda x: x.to_dict(orient='records'))
dfGrouped = dfGrouped.reset_index()
```

```python
df3 = df2.join(dfGrouped, lsuffix='farm_id', rsuffix='ARS_FARM_ID')
del df3['ARS_FARM_ID']
print df3
```

```python
df3 = df3.rename(columns={0:'fields'})
```

```python
print df3
```

```python
df4 = df.filter(['GrowerName', 'ARS_GROWER_ID', 'BRANCH_NAME', 'DIVISION_NAME', 'REGION_NAME'], axis=1)

df4 = df4.rename(columns={'ARS_GROWER_ID':'grower_id'})

dfTemp = df.filter(['ARS_GROWER_ID', 'ARS_FARM_ID'])

dfTemp2 = dfTemp.join(df3, lsuffix='ARS_FARM_ID', rsuffix='farm_id')

del dfTemp2['ARS_FARM_ID']
```

```python
print dfTemp2 
```

```python
dfGrouped2 = dfTemp2.groupby(['ARS_GROWER_ID']).apply(lambda x: x.to_dict(orient='records'))
dfGrouped2 = dfGrouped2.reset_index()
dfGrouped2 = dfGrouped2.rename(columns={0:'farms'})
```

```python
df5 = df4.join(dfGrouped2, lsuffix='grower_id', rsuffix='ARS_GROWER_ID')
del df5['ARS_GROWER_ID']
print df5
```

```python
df5 = df5.rename(columns={0:'farms'})
```

```python
print df5
```

```python
for index, row in df5.iterrows():
    branchName = str(row['BRANCH_NAME']).strip()
    regionName = str(row['REGION_NAME']).strip()
    divisionName = str(row['DIVISION_NAME']).strip()
    growerId = str(row['grower_id']).strip()
    
    fileName =  regionName + '/' + divisionName + '/' + branchName + '/' + growerId + '.json'
    jsonData =  row.to_json()

    try:
        os.makedirs(os.path.dirname(fileName))
    except:
        pass
   
    try:
        obj = None
        obj = open(fileName, 'w+')
        obj.write(jsonData)
        obj.close
    except:
        print "Something went wrong"
    
    
```
