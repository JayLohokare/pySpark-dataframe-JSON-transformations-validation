```python
farmsJSONified = farms.apply(lambda x: x.to_dict(orient='records'))
farmsJSONified = farmsJSONified.reset_index()
farmsJSONified = farmsJSONified.rename(columns={0:'farms'})
print farmsJSONified
```

```python
for index, row in farmsJSONified.iterrows():
    branchName = str(row['BRANCH_NAME']).strip()
    regionName = str(row['REGION_NAME']).strip()
    divisionName = str(row['DIVISION_NAME']).strip()
    growerId = str(row['ARS_GROWER_ID']).strip()
    
    fileName =  regionName + '/' + divisionName + '/' + branchName + '/' + growerId + '.json'
    jsonData =  row.to_json()

    try:
        os.makedirs(os.path.dirname(fileName))
    except:
        pass
   
    try:
        obj = None
        obj = open(fileName, 'w+')
        obj.write(jsonData)
        obj.close
    except:
        print "Something went wrong"
    
    
```
