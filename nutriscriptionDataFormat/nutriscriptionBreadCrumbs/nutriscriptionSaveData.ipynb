```python
import pandas as pd
from collections import defaultdict
import json
import os
```

```python
df = pd.read_csv("all_fields_with_Region.csv", '|')
```

```python
newTemp = df.groupby([ 'DIVISION_ID', 'ARS_GROWER_ID', 'BRANCH_ID'])
len(newTemp)
```

```python
df2 = df.filter(['FarmName', 'ARS_FARM_ID', 'ARS_GROWER_ID', 'BRANCH_ID', 'DIVISION_ID'], axis=1)
df2 = df2.rename(columns={'ARS_FARM_ID':'farm_id', 'ARS_GROWER_ID': 'grower_id', 'BRANCH_ID' : 'branch_id', 'DIVISION_ID':'division_id'})
```

```python
dfGrouped = df.filter(['ID', 'BRANCH_ID', 'DIVISION_ID',  'ARS_FIELD_NAME', 'ARS_FARM_ID' , 'ARS_GROWER_ID']).groupby([ 'ARS_FARM_ID','DIVISION_ID', 'ARS_GROWER_ID', 'BRANCH_ID']).apply(lambda x: x.to_dict(orient='records'))
dfGrouped = dfGrouped.reset_index()
print dfGrouped 
```

```python
df3 = dfGrouped.join(df2, rsuffix=['branch_id','division_id','farm_id', 'grower_id'], lsuffix=['BRANCH_ID', 'DIVISION_ID',  'ARS_FARM_ID','ARS_GROWER_ID'])
del df3['ARS_FARM_ID']
del df3['ARS_GROWER_ID']
del df3['BRANCH_ID']
del df3['DIVISION_ID']
df3 = df3.rename(columns={0:'fields'})
list(df3)
print df3
```

```python
# sum(pd.isnull(df3['fields']))
```

```python
df3 = df3.rename(columns={'farm_id':'ARS_FARM_ID', 'grower_id':'ARS_GROWER_ID', 'branch_id': 'BRANCH_ID',  'division_id':'DIVISION_ID'})
print df3
```

```python
dfTemp2 = df3.groupby(['ARS_GROWER_ID', 'DIVISION_ID', 'BRANCH_ID']).apply(lambda x: x.to_dict(orient='records'))
dfTemp2 = dfTemp2.reset_index()
dfTemp2 = dfTemp2.rename(columns={0:'farms'})

# print dfTemp2 

print df3['DIVISION_ID'].unique()
```

```python
df4 = df.filter(['GrowerName', 'ARS_GROWER_ID', 'BRANCH_ID','BRANCH_NAME', 'DIVISION_NAME', 'REGION_NAME'], axis=1)
df4 = df4.rename(columns={'ARS_GROWER_ID':'grower_id', 'BRANCH_ID':'branch_id'})
print df4 
```

```python
final = dfTemp2.join(df4, rsuffix=['grower_id','branch_id'], lsuffix=['ARS_GROWER_ID','BRANCH_ID'])
del final['grower_id']
del final['branch_id']
print final
```

```python
len(final)
```

```python
for index, row in final.iterrows():
    branchName = str(row['BRANCH_NAME']).strip()
    regionName = str(row['REGION_NAME']).strip()
    divisionName = str(row['DIVISION_NAME']).strip()
    growerId = str(row['ARS_GROWER_ID']).strip()
    
    fileName =  regionName + '/' + divisionName + '/' + branchName + '/' + growerId + '.json'
    jsonData =  row.to_json()

    try:
        os.makedirs(os.path.dirname(fileName))
    except:
        pass
   
    try:
        obj = None
        obj = open(fileName, 'w+')
        obj.write(jsonData)
        obj.close
    except:
        print "Something went wrong"
    
    
```
